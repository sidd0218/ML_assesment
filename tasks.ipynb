```python
ty = pd.DataFrame(data.groupby(['INSURANCE_TYPE'])['PREMIUM_AMOUNT'].sum())
ty['INSURANCE_TYPE'] = ty.index
ty.reset_index(drop=True,inplace=True)
ty
```

```python
data['PREMIUM_AMOUNT'].sum()
```

```python
ty['PERCENTAGE'] = (ty['PREMIUM_AMOUNT']/data['PREMIUM_AMOUNT'].sum())*100
```

```python
def pre(row):
    global val
    if((row['INSURANCE_TYPE'] == 'Health') | (row['INSURANCE_TYPE'] == 'property')):
        val = row['PREMIUM_AMOUNT'] + (0.07* row['PREMIUM_AMOUNT'])
    elif((row['INSURANCE_TYPE'] == 'Life') | (row['INSURANCE_TYPE'] == 'Motor')):
        val = row['PREMIUM_AMOUNT'] + (0.02* row['PREMIUM_AMOUNT'])
    elif((row['INSURANCE_TYPE'] == 'Mobile') | (row['INSURANCE_TYPE'] == 'Travel')):
        val = row['PREMIUM_AMOUNT'] - (0.10* row['PREMIUM_AMOUNT'])
    return val
ty['NEW_PREMIUM'] = ty.apply(pre , axis= 1)
```

```python
ty['NEW_PREMIUM_PERCENTAGE'] = ty['NEW_PREMIUM']/(ty['NEW_PREMIUM'].sum())*100
ty.head()
```

```python
#task9
def premium (data):
    global n
    if((data['TENURE'] > 60) and (data['EMPLOYMENT_STATUS'] == 'N') and (data['NO_OF_FAMILY_MEMBERS'] >=4)):
        n =1
    else:
        n=0
    return n
```

```python
data['ELIGIBLE_FOR_DISCOUNT'] = data.apply(premium, axis=1)
data['ELIGIBLE_FOR_DISCOUNT'].mean()
```
